```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd
import numpy as np

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
df = pd.read_csv("data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv")
df.head()
```

```python
# Watch out for leap days (i.e. February 29th), 
#it is reasonable to remove these points from 
#the dataset for the purpose of this visualization.

# get year and month-date from Date column 
df['Year'], df['Month-Date'] = zip(*df['Date'].apply(lambda x: (x[:4], x[5:])))
# remove leap years from data set
df = df[(df['Month-Date'] != '02-29')]
```

```python
# Overlay a scatter of the 2015 data for any points (highs and lows) for which 
# the ten year record (2005-2014) record high or record low was broken in 2015.

# record low
temp_min = df[(df['Element'] == 'TMIN') & (df['Year'] != '2015')].groupby('Month-Date').aggregate({'Data_Value':np.min})
# record high
temp_max = df[(df['Element'] == 'TMAX') & (df['Year'] != '2015')].groupby('Month-Date').aggregate({'Data_Value':np.max})

# extra copies of record low and record high of 2015 to compare for broken min and max
compare_min = df[(df['Element'] == 'TMIN') & (df['Year'] == '2015')].groupby('Month-Date').aggregate({'Data_Value':np.min})
compare_max = df[(df['Element'] == 'TMAX') & (df['Year'] == '2015')].groupby('Month-Date').aggregate({'Data_Value':np.max})
# find broken min and max from 2016
broken_min = np.where(compare_min['Data_Value'] < temp_min['Data_Value'])[0]
broken_max = np.where(compare_max['Data_Value'] > temp_max['Data_Value'])[0]
```

```python
plt.figure()
plt.title("Temperature Summary Plot Near Ann Arbor, Michigan, United States")
plt.xlabel("Day of the Year")
plt.ylabel("Temperature (Tenths of degrees C)")
plt.plot(temp_min.values, 'b', label = 'Record Low')
plt.plot(temp_max.values, 'r', label = 'Record High')
plt.scatter(broken_min, compare_min.iloc[broken_min], s = 10, c = 'g', label = 'broken low')
plt.scatter(broken_max, compare_max.iloc[broken_max], s = 10, c = 'm', label = 'broken high')
plt.gca().axis([-5, 370, -400, 650])
plt.xticks(range(0, len(temp_min), 20), temp_min.index[range(0, len(temp_min), 20)], rotation = '45')
plt.legend(loc = 1, frameon = False)
plt.gca().fill_between(range(len(temp_min)), temp_min['Data_Value'], temp_max['Data_Value'], facecolor = 'yellow', alpha = 0.5)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.show()
```
